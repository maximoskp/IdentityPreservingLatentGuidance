```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
os.makedirs('figs', exist_ok=True)
# Increase font sizes
plt.rcParams.update({'font.size': 18})
```

```python
se_pre = pd.read_csv('results/SE/pretraining.csv')
se_pre.head()
```

```python
# Normalize n_vis to [0,1]
n_vis_norm = (se_pre['n_vis'] - se_pre['n_vis'].min()) / (se_pre['n_vis'].max() - se_pre['n_vis'].min())

# val_acc is already in [0,1]
val_acc = se_pre['val_acc']

# Normalize val_ppl to [0,1]
val_ppl_norm = (se_pre['val_ppl'] - se_pre['val_ppl'].min()) / (se_pre['val_ppl'].max() - se_pre['val_ppl'].min())

# Plot
plt.figure(figsize=(10, 6))
plt.plot(se_pre['epoch'], n_vis_norm, color='grey', alpha=0.5, label='n_vis (norm.)')
plt.plot(se_pre['epoch'], val_acc, color='red', label='val_acc')
plt.plot(se_pre['epoch'], val_ppl_norm, color='green', label='val_ppl (norm.)')
plt.plot([203,203], [0, 1], color='blue', linestyle='--', label='Epoch 203 (best)')

plt.xlabel('epoch')
plt.ylabel('Normalized Value')
plt.ylim(0, 1)
# plt.title('Training Metrics Over Epochs')
plt.legend()
plt.grid(True)
plt.savefig('figs/pre_se_metrics.png')
plt.show()
```
